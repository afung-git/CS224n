```python
from util import eval_dicts
import json
```

```python
gold = {}
with open('data/dev-v2.0.json') as target_json:  
    targets = json.load(target_json)
    for article in targets["data"]:
        for para in article["paragraphs"]:
            for qa in para["qas"]:
                gold.update({qa["id"]: {"answers": [ans['text'] for ans in qa["answers"]]}})
```

```python
with open('../bert/gugel_bert/predictions.json') as pred_json:
    preds = json.load(pred_json)
    results = eval_dicts(gold, preds, True)
```
